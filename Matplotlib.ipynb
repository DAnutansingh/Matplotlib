```python
#Bar plot
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscube",fontsize=15)
c=['y','b','m','g']
plt.bar(x,y,width=0.4,color="r",align="edge",edgecolor="r",linewidth=5,linestyle=":",alpha=0.5,label="popularity")
plt.bar(x,z,width=0.4,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscube",fontsize=15)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
#Matplotlib Stack Plot
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","y"],baseline="zero")

plt.title("python")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc=2)
plt.show()
```

```python
#Matplotlib Step Plot

import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,2,3,5,4]

plt.step(x,y,marker="o",color="r",ms=10,mfc="g",label="python")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
#Fill_Between Plot in Matplotlib

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.5)

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
